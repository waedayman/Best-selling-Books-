**Dataset Introduction:** This dataset scraped from Wikipedia's List of best selling books page. "Best-selling" refers to the estimated number of copies sold of each book, rather than the number of books printed or currently owned.
The dataset under consideration provides a comprehensive overview of the best-selling books, capturing crucial details that contribute to their success. The dataset includes a range of information about each book, facilitating an exploration of various aspects related to their popularity and impact.


---







**Key Attributes:**

**1.Books:** This column contains the titles of the best-selling books, providing insight into the literary works that have achieved remarkable success.

**2.Authors:** The authors' names are listed, allowing for an examination of the individuals behind the best-selling books.

**3.Language:** The dataset includes information on the languages in which the books are published, shedding light on linguistic diversity among best-sellers.

**4.First_Published:** This attribute specifies the year of the first publication for each book, enabling the analysis of temporal trends in best-selling literature.

**5.Sales_in_millions:** The sales figures, measured in millions, represent the estimated number of copies sold for each book. This attribute serves as a quantitative measure of a book's commercial success.

**Import Needed Libraries**

```python
import pandas as pd
import plotly.express as px
from plotly.offline import iplot
```

**Reading Data And Getting Info**

```python
df = pd.read_csv('books_data.csv')
df.head()
```

**Data Cleaning and Additional Information**

```python
df.isnull().sum()
df = df.dropna()
df = df.drop('Unnamed: 0', axis=1, errors='ignore')
df = df.replace('–present', '', regex=True)

pd.DataFrame({
    'Count': df.count(),
    'Null': df.isnull().sum(),
    'Cardinality': df.nunique()
})
```

**Top 10 Best Books**

```python
df_10_best_books = df[['Books', 'Sales_in_millions']].sort_values(by='Sales_in_millions', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(df_10_best_books['Books'], df_10_best_books['Sales_in_millions'])
plt.xlabel('Title')
plt.ylabel('Sales in millions')
plt.title('Top 10 Best Books')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

A bar chart showcased the top 10 best-selling books based on sales in millions.

---



**Top 10 Authors with Multiple Big Hits**

```python
df_best_authors = df['Authors'].value_counts().head(10)
plt.figure(figsize=(10, 6))
df_best_authors.plot(kind='bar', title='Top 10 Authors with Multiple Big Hits')
plt.xlabel('Authors')
plt.ylabel('Number of Big Hits')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

A bar chart highlighted authors with multiple best-selling books, emphasizing their impact on the literary world.

---



**Total Sales Over Time For Most Popular Languages**

```python
language_sales = df.groupby('Language')['Sales_in_millions'].sum().reset_index()
plt.figure(figsize=(12, 8))
plt.plot(language_sales['Language'], language_sales['Sales_in_millions'], marker='o', linestyle='-', color='b')
plt.title('Total Sales Over Time for Most Popular Languages')
plt.xlabel('Language')
plt.ylabel('Total Sales in millions')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

A line graph illustrated the total sales over time for the most popular languages.

It provided insights into language preferences among readers.

---



**Best Year Of Publications**

```python
df_best_year = df['First_Published'].value_counts().head(20)
plt.figure(figsize=(12, 6))
df_best_year.plot(kind='bar', title='Best Year of Publications')
plt.xlabel('Year of Publication')
plt.ylabel('Number of Publications')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

A bar chart visualized the years with the highest number of best-selling publications.

This analysis uncovered trends and patterns in the popularity of books over time.
